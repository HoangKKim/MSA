```python
#import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data from covid-19 cases
df = pd.read_csv('data/covid-19-cases.csv')
df
```

```python
# group by country column
groupedCountries = df.groupby("Country/Region")

# Sum over the different states/regions of each country
df_countries = groupedCountries.sum()
df_countries
```

```python
# Plot the first graph - line graph
def visualizeCasesInCountry(country):
    # extract the series corresponding to the case numbers for country
    country_df = df_countries.loc[country, df_countries.columns[4:]]

    # convert the index to proper datetime object
    country_df.index = pd.to_datetime(country_df.index)

    # get the number of column
    n = len(country_df)
    
    # setting bar with corresponding value
    plt.plot(range(n), country_df.values, '#5730C2')

    # set label for X, Y axis
    plt.xlabel('Days')
    plt.ylabel('Confirmed cases, $N$')

    # set title for this graph 
    title = '{}\n{} cases on {}'.format(country, country_df[-1],country_df.index[-1].strftime('%d %B %Y'))
    plt.title(title)
    
    plt.show()
    
visualizeCasesInCountry("China")
```

```python
from bokeh.plotting import figure, show
from bokeh.layouts import column
from bokeh.models import ColumnDataSource, Select, NumeralTickFormatter
from bokeh.io import curdoc
import pandas as pd

# Đọc file csv và nhóm theo cột "Country/Region" (các giá trị sum lại với nhau)
df = pd.read_csv("./covid-19-cases.csv").groupby("Country/Region").sum().iloc[:, 4:]

# Chuyển đổi cột thành datetime
df.columns = pd.to_datetime(df.columns)

# Các quốc gia ở Đông Nam Á
asean_countries = [
    'Brunei', 'Cambodia', 'East Timor', 'Indonesia', 'Laos', 
    'Malaysia', 'Myanmar', 'Philippines', 'Singapore', 'Thailand', 'Vietnam'
]

# Lọc và tạo biểu đồ với các quốc gia ở ĐNA
df_asean = df[df.index.isin(asean_countries)]

# Chuẩn bị dataset
source = ColumnDataSource(data=dict(date=df_asean.columns, cases=df_asean.loc["Vietnam"].values))

# Tạo figure
p = figure(title="COVID-19 Cases", x_axis_label='Date', y_axis_label='Cases', x_axis_type='datetime', width=800, height=400)
p.line(x='date', y='cases', source=source, line_width=2)
p.yaxis[0].formatter = NumeralTickFormatter(format="0,0")

# Tạo Select widget
select = Select(title="Select Country", value="Vietnam", options=list(df_asean.index))

# Hàm update data
def update_plot(attr, old, new):
    selected_country = select.value
    new_data = dict(date=df_asean.columns, cases=df_asean.loc[selected_country].values)
    source.data = new_data
    p.title.text = f"COVID-19 Cases in {selected_country}"

# Link widget với hàm update
select.on_change('value', update_plot)

# Sắp xếp layout
layout = column(select, p)

# Thêm layout vào tài liệu hiện tại của Bokeh server
curdoc().add_root(layout)
```

```python
# Plot the second graph - bar graph
def visualizeTopCountryInDay(date):
    top5_df = df_countries.sort_values(by= date, ascending=False).head()
    labels = top5_df.index
    
    # convert to datetime
    d = pd.to_datetime(date)
    
    # set characteristic for the graph
    plt.bar(labels, top5_df[date], color = "#628eec")
    
    # write down specific figures
    for i, value in enumerate(top5_df[date]):
        plt.text(i, value, str(value), ha='center', va='bottom', fontsize = 8)
        
    # set label for X, Y axis
    plt.xlabel("Country/Region")
    plt.ylabel("Confirmed cases, $N$")
    
    # set title for this graph
    title = "Countries/Region with highest Covid-19 cases \n on {}".format(d.strftime('%d %B %Y'))
    plt.title(title, fontsize=12, pad =  20, va = "bottom")
    
    plt.show()
    
visualizeTopCountryInDay("7/14/21")
```

```python
# Plot the third graph - pie graph
def visualizeCasesInStatesOfCountry(country, threshold): 
    df_subset = df[df["Country/Region"]== country]
    
    # check the state 
    if(len(df_subset) == 1):
        print("Invalid")
        return
    
    # reset index for df_subset
    df_subset = df_subset.set_index("Province/State")
    
    # compute sum of cases in each state/province
    sum_rows = df_subset.iloc[:, 4:].sum(axis=1)

    # apply threshold to filter great and small cases
    great_cases = sum_rows[sum_rows>=threshold]
    small_cases = sum_rows[sum_rows<threshold]
    
    # sum value of element in small_cases to make new component "Other"
    sum_small_cases = pd.Series([small_cases.sum()], index = ["Other"])

    # combine great_case and "Other" 
    df_final = pd.concat([great_cases, sum_small_cases])
    
    # set characteristics of the graph
    plt.pie(df_final.values, labels = df_final.index, autopct='%1.1f%%', labeldistance=1.2)
    # present labels in a table label on the graph
    plt.legend()
    # set the title for this graph
    plt.title("Covid-19 cases of states/provinces of {}".format(country))
    plt.show()

visualizeCasesInStatesOfCountry("China", 700_000)

    
```
